```python
def hitung_nilai_akhir(uts, uas):
    nilai_akhir = (uts + uas) / 2
    return nilai_akhir

uts = float(input("input nilai UTS: "))
uas = float(input("input nilai UAS: "))

nilai_akhir = hitung_nilai_akhir(uts, uas)

if nilai_akhir >= 70:
    print("mahasiswa lulus dengan nilai akhir", nilai_akhir)
else:
    print("mahasiswa tidak lulus dengan nilai akhir", nilai_akhir)
```

```python
print ("bilangan ganjil dari 1 sampai 50: ")

for i in range(1,51):
    if i % 2 != 0:
        print(i, end=" ")

print ("\nbilangan genap dari 1 sampai 50: ")

for j in range(1,51):
    if j % 2 == 0:
        print(j, end=" ")
```

```python
def square(n):
    return n ** 2

numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)

print(list(squared_numbers))
```

```python
def is_even(n):
    return n % 2 == 0

numbers = [1 ,2, 3, 4, 5, 6]
even_numbers = filter(is_even, numbers)

print(list(even_numbers))
```
