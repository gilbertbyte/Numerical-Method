```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f(x) = 1/2 * x * e^(x^(0.5)) - 2
def f(x):
    return 0.5 * x * np.exp(x**0.5)-2
```

```python
# f'(x) = 0.5 * e^(akar x) * (1 + 0.5 * akar x)
def turunan(x):
    return(0.5* np.exp(np.sqrt(x))) * (1 + 0.5 * np.sqrt(x))
```

```python
# x1 itu xi, es itu error toleransi
def newtonRaphson(x1, es):
    # inisialisasi variabel
    ea = 100
    iterasi = 0
    list_ea = []
    list_xr = []
    list_x = []

    # loop selama error lebih besar dari error toleransi
    while (ea > es):
        # iterasi tambahv1
        iterasi += 1
        print(f"iterasi ke-{iterasi}")

        # rumus newton raphson
        # x2 itu xr, x1 itu xi
        x2 = x1 - f(x1)/turunan(x1)

        # masukkan nilai iterasi dan xr ke dalam list
        list_x.append(iterasi)
        list_xr.append(x2)

        # print nilai xi, xr, f(xi), f'(xi), \t itu tab, :.4f itu 4 angka di belakang koma
        print(f"xi \t= {x1:.4f}")
        print(f"xi+1 \t= {x2:.4f}")
        print(f"f(xi) \t= {f(x1):.4f}")
        print(f"f'(xi) \t= {turunan(x1):.4f}")

        # kalau iterasi lebih dari 1, hitung ea
        if(iterasi > 1):
            ea = abs((x2-x1)/x2)*100
            list_ea.append(ea)
            print(f"ea \t= {ea:.4f}")

        # xi di ganti jadi xi+1 untuk iterasi selanjutnya
        x1 = x2

    return list_ea, list_xr, list_x
```

```python
print("Newton Raphson")
# panggil fungsi newtonRaphson dengan argumen xi = 1, es = 4 (sesuai teori)
ea, xr, x = newtonRaphson(1, 4)
```

```python
# CATATAN ALUR FUNCTION NEWTON RAPHSON (tidak di run)
# ITERASI 2

def newtonRaphson(x1, es):
    ea = 100
    iterasi = 0
    list_ea = []
    list_xr = []
    list_x = []

    # 13. masih ea=100 > es=4, masuk ke loop
    while (ea > es):
        # 14. sblmnya iterasi=1, jadi 1 karena iterasi=iterasi+1 -> iterasi=1+1=2
        iterasi += 1
        print(f"iterasi ke-{iterasi}")

        # 15. xr = 1.2868
        x2 = x1 - f(x1)/turunan(x1)

        # 16. list_x = [1, 2]
        list_x.append(iterasi)
        # 17. list_xr = [1.3143, 1.2868]
        list_xr.append(x2)

        # 18. xi=1.3143, xr=1.2868, f(xi)=0.681, f'(xi)=2.4755
        print(f"xi \t= {x1:.4f}")
        print(f"xi+1 \t= {x2:.4f}")
        print(f"f(xi) \t= {f(x1):.4f}")
        print(f"f'(xi) \t= {turunan(x1):.4f}")

        # 11. iterasinya sudah 2, hitung ea
        if(iterasi > 1):
            # 12. ea = 2.1385%
            ea = abs((x2-x1)/x2)*100
            # 13. list_ea = [2.1385]
            list_ea.append(ea)
            print(f"ea \t= {ea:.4f}")

        # 14. x1=1.2868
        x1 = x2

    # 15. return list_ea = [2.1385], list_xr = [1.3143, 1.2868], list_x = [1, 2]
    return list_ea, list_xr, list_x
```
