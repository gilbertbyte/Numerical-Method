```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f(x) = 1/2 * x * e^(x^(0.5)) - 2
def f(x):
    return 0.5 * x * np.exp(x**0.5)-2
```

- x2 itu xi+1 (xr)
- x1 itu xi-1
- x0 itu xi

- Dari soal teori xi-1 = 1 dan xi = 2

```python
# parameternya x0(xi-1)=2 (xi), x1 (xi-1)=1, es=0.04
def Secant(x0, x1, es):
    # inisialisasi variabel
    ea = 100
    iterasi = 0
    nilai_ea = []
    nilai_xr = []
    nilai_x = []

    # loop selama ea > es
    while (ea > es):
        # iterasi bertambah
        iterasi += 1
        print(f"iterasi ke-{iterasi}")

        # hitung xr dengan rumus secant
        x2 = x0 - (f(x0)*(x0-x1))/(f(x0)-f(x1))

        # tambahkan iterasi ke list_x, dan xr ke list_xr
        nilai_x.append(iterasi)
        nilai_xr.append(x2)

        # tampilkan xi, xi-1, xi+1, f(xi), f(xi-1)
        print(f"xi \t= {x0:.4f}")           
        print(f"xi-1 \t= {x1:.4f}")
        print(f"xi+1 \t= {x2:.4f}")
        print(f"f(xi) \t= {f(x0):.4f}")
        print(f"f(xi-1) = {f(x1):.4f}")

        # hitung ea jika iterasi > 1
        if(iterasi > 1):
            ea = abs((x2-x0)/x2)*100
            nilai_ea.append(ea)
            
            print(f"ea \t= {ea:.4f}")

        x1 = x0 # xi-1 iterasi berikutnya = xi iterasi sebelumnya
        x0 = x2 # xi iterasi berikutnya = xi+1 iterasi sebelumnya
        
    return nilai_ea, nilai_xr, nilai_x
```

```python
print("Secant")
ea, xr, x = Secant(2,1,4) # argumen fungsi x0=2 (xi), x1=1 (xi-1), es=0.04
```

```python
# subplot 1
plt.subplot(211)
plt.title("Secant")
plt.plot(x[1:], ea, "--o")
plt.xlabel("Loop ke berapa")
plt.ylabel("nilai error")
plt.xticks(x[1:])

# subplot 2
plt.subplot(212)
plt.plot(x, xr, "--o")
plt.xlabel("Loop ke berapa")
plt.ylabel("nilai xr")
plt.xticks(x)
plt.subplots_adjust(hspace=0.3)
plt.savefig("Secant.png")
plt.show()
```
