```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f(x) = 1/2 * x * e^(x^(0.5)) - 2
def f(x):
    return 0.5 * x * np.exp(x**0.5)-2
```

```python
# f'(x) = 0.5 * e^(akar x) * (1 + 0.5 * akar x)
def turunan(x):
    return(0.5* np.exp(np.sqrt(x))) * (1 + 0.5 * np.sqrt(x))
```

```python
# x1 itu xi, es itu error toleransi
def newtonRaphson(x1, es):
    # inisialisasi variabel
    ea = 100
    iterasi = 0
    list_ea = []
    list_xr = []
    list_x = []

    # loop selama error lebih besar dari error toleransi
    while (ea > es):
        # iterasi tambahv1
        iterasi += 1
        print(f"iterasi ke-{iterasi}")

        # rumus newton raphson
        # x2 = xi+1 = xr, x1 itu xi
        x2 = x1 - f(x1)/turunan(x1)

        # masukkan nilai iterasi dan xr ke dalam list
        list_x.append(iterasi)
        list_xr.append(x2)

        # print nilai xi, xr, f(xi), f'(xi), \t itu tab, :.4f itu 4 angka di belakang koma
        print(f"xi \t= {x1:.4f}")
        print(f"xi+1 \t= {x2:.4f}")
        print(f"f(xi) \t= {f(x1):.4f}")
        print(f"f'(xi) \t= {turunan(x1):.4f}")

        # kalau iterasi lebih dari 1, hitung ea
        if(iterasi > 1):
            ea = abs((x2-x1)/x2)*100
            list_ea.append(ea)
            print(f"ea \t= {ea:.4f}")

        # xi di ganti jadi xi+1 untuk iterasi selanjutnya
        x1 = x2

    return list_ea, list_xr, list_x
```

```python
print("Newton Raphson")
# panggil fungsi newtonRaphson dengan argumen xi = 1, es = 0.0001
ea, xr, x = newtonRaphson(1,0.0001)
```

```python
# subplot 1
plt.subplot(211)
plt.title("Newton Raphson")
plt.plot(x[1:], ea, "--o")
plt.xlabel("Loop ke berapa")
plt.ylabel("nilai error")
plt.xticks(x[1:])

# subplot 2
plt.subplot(212)
plt.plot(x, xr, "--o")
plt.xlabel("Loop ke berapa")
plt.ylabel("nilai xr")
plt.xticks(x)
plt.subplots_adjust(hspace=0.3)
plt.savefig("NewtonRaphson.png")
plt.show()
```
