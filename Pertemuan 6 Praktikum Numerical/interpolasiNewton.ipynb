```python
def error(Ns, Ne):
    return abs(Ns-Ne) / Ns * 100
```

```python
# X itu nilai yang mau dicari, Ns nilai sebenarnya
def ordeSatu(X, x0, x1, fx0, fx1, Ns):
    # Ne itu nilai estimasi atau f1(x) kalau dalam teori
    Ne = (fx0 + ((fx1 - fx0) / (x1 - x0)) * (X - x0))

    print("Interpolasi Newton Orde Satu:")
    print("Nilai Estimasi: %.4f" %(Ne))
    print("Error / Kesalahan Sebenarnya: %.4f" %(error(Ns, Ne)))
```

```python
ordeSatu(11.5, 10.5, 12, 45.2, 58.1, 55.92)
```

```python
def ordeDua(X, x, fx, Ns):
    # inisialisasi list b
    b = []
    # b0 = f(x0)
    b.append(fx[0]) # b = [b0]

    for i in range(1, len(x)):
        if(i ==  1):
            # b1 = (fx[1] - fx[0]) / (x[1] -x[0])
            B = (fx[i] - fx[i-1]) / (x[i] -x[0])
            b.append(B) # b = [b0, b1]
        elif(i == 2):
            #b2=(((fx[2] - fx[1]) / (x[2] - x[1])) - b1]) / (x[i] -x[0])
            B = (((fx[i] - fx[i-1]) / (x[i] - x[i-1])) - b[i-1]) / (x[i] -x[0])
            b.append(B) # b = [b0, b1, b2]
    
    # nilai estimasi awalnya di inisialiasi dengan b0
    Ne = b[0]

    # loop untuk menambahkan suku polinom 
    for i in range(1, len(x)):
        # temp untuk menyimpan hasil perkalian, awalnya 1
        temp = 1
        # membentuk (x - x0)(x - x1)... sesuai orde
        for j in range(0, i):
            # contoh jika iterasinya 0 maka temp = 1 * (x - x0)
            temp = temp * (X-x[j])

        # contoh iterasi 1
        # Ne = Ne + b[i] * temp
        # Ne = b[0] + b[1] * (x - x[0])
        Ne += b[i] * temp
    
    print("Interpolasi Newton Orde Dua:")
    print("b0 = %.4f, b1 = %.4f, b2 = %.4f" %(b[0], b[1], b[2]))
    print("Nilai Estimasi: %.4f" %(Ne))
    print("Error / Kesalahan Sebenarnya: %.4f" %(error(Ns, Ne)))
```

```python
x = [10.5, 12, 12.4]
fx = [45.2, 58.1, 60.2]
ordeDua(11.5, x, fx, 55.92)
```

```python
def ordeTiga(X, x, fx, Ns):
    # jumlah titik data 
    n = len(x)

    # bikin list kosong b
    b = [0] * n

    # isi awal b dengan f(x)
    # b = [f(x0), f(x1), f(x2), f(x3)]
    for i in range(n):
        b[i] = fx[i]

    # loop dari 1 sampai n-1 (jumlah orde)
    for j in range(1, n):
        # loop mundur, mulai dari indeks terakhir (n-1), berhenti di j
        for i in range(n-1, j-1, -1):
            # rumus b yang bagian awal
            b[i] = (b[i] - b[i-1]) / (x[i] - x[i-j])

    # rumus aslinya f(x)=b0​+b1​(x−x0​)+b2​(x−x0​)(x−x1​)+b3​(x−x0​)(x−x1​)(x−x2​)
    # ubah jd f(x)=b0​+(x−x0​)(b1​+(x−x1​)(b2​+(x−x2​)b3​))
    Ne = b[n-1]
    # loop mundur dari indeks kedua terakhir sampai ke awal
    for i in range(n-2, -1, -1):
        # rumus baru td
        # (X - x[i]) dikalikan hasil sebelumnya
        # lalu ditambah koefisien b[i]
        Ne = b[i] + (X-x[i]) * Ne
    
    print("Interpolasi Newton Orde Tiga:")
    for i in range(i, n):
        print("b%d = %.4f" %(i, b[i]))
    print("Nilai Estimasi: %.4f" %(Ne))
    print("Error / Kesalahan Sebenarnya: %.4f" %(error(Ns, Ne)))
```
