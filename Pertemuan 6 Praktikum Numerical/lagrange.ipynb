```python
import numpy as np

# fungsi error
# ne = nilai eksak (nilai sebenarnya)
# ns = nilai hasil interpolasi (hasil hitungan)
def e (ne, ns):
    return abs((ns-ne)/ns * 100)

# fungsi lagrange
# x -> data, 
# fx -> f(x), 
# orde -> derajat polinom, 
# x_baru -> nilai yang mau ditaksir, 
# es -> nilai sbnrnya (buat error) 
def lagrange(x, fx, orde, x_baru, es):
    # list kosong buat simpan titik yang dipakai untuk interpolasi
    # karena orde n cuma butuh n+1 titik data
    x_temp = []
    fx_temp = []

    # loop buat cari posisi x_baru
    for i in range(len(x)):
        # kalau xi ​< x_baru ​< xi+1
        if (x[i] < x_baru and x[i+1] > x_baru):
            # ambil jumlah titik sesuai orde
            for j in range(orde + 1):
                # simpan titik yang diambil ke list s
                x_temp.append(x[i+j])
                fx_temp.append(fx[i+j])
            break
    
    # inisialisasi L
    L = np.ones(orde+1)

    # hitung L
    # loop untuk setiap i, buat hitung L1, L2, dst
    for i in range(orde+1):
        print("L%d =  " %(i))
        # untuk setiap i, loop semua j yg bukan i
        for j in range(orde+1):
            # skip kalau i == j
            if (i != j):
                print("x - x%d / x%d - x%d " %(j, i, j))
                # (x - xj) / (xi - xj)
                temp = ((x_baru - x_temp[j]) / (x_temp[i] - x_temp[j]))
                # L[i] = L[i] * temp
                L[i] *= temp
        print("L%d = %.4f" %(i, L[i]))
        print()

    sum = 0
    for i in range(orde + 1):
        # hitung f(X) = L[i] * fx_temp[i]
        sum += (L[i] * fx_temp[i])

    print("F%d(x) = %f" %(orde, sum))
    print("error : %.4f" %(e(es, sum)))
```

```python
x = [8.8, 9.6, 10.5, 12, 12.4, 12.8, 13.2]
fx = [30.8, 37.1, 45.2, 58.1, 60.2, 64.4, 68.8]
orde = 3
es = 55.92
X = 11.5

lagrange(x, fx, orde, X, es)
```
