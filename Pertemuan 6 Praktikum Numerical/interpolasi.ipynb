```python
import numpy as np
```

```python
data = [
    {'x': 6.5079, 'y': 16.4100},
    {'x': 8.0814, 'y': 16.6266},
    {'x': 3.9090, 'y': 15.9003},
    {'x': 6.1083, 'y': 16.3467},
    {'x': 9.9295, 'y': 16.8325},
    {'x': 9.9629, 'y': 16.8359},
    {'x': 2.2559, 'y': 15.3505},
    {'x': 3.0724, 'y': 15.6595},
    {'x': 1.5147, 'y': 14.9522},
    {'x': 5.4081, 'y': 16.2249}
]

Xp = 17.51321 
data.sort(key=lambda p: abs(p["x"] - Xp))
orde = 5
X = np.array([p["x"] for p in np.array(data)])
Y = np.array([p["y"] for p in np.array(data)])

def interpolasi(Xp, X, Y, orde):
    b = np.zeros((orde + 1, orde + 1))
    b[:, 0] = Y[:orde + 1]
    for j in range(1, orde + 1):
        for i in range(orde - j + 1):
            b[i,j] = (b[i+1,j-1] - b[i,j-1]) / (X[i+j] - X[i])
    Xt = 1
    Yp = b[0,0]
    for j in range(orde):
        Xt = Xt * (Xp - X[j])
        Yp = Yp + b[0,j+1] * Xt

    print ("f(x = %.2f) = %f (newton orde %d)" % (Xp, Yp, orde))

interpolasi(Xp, X, Y, orde)
```
