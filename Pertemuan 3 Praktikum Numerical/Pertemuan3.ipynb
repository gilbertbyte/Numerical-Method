```python
x = np.array([[1,2],[3,4]], dtype = float)
y = np.array([[5,6],[7,8]], dtype = float)

print(x)
print(y)
```

```python
print("penjumlahan")
print(x+y)
print(np.add(x,y))
```

```python
print("pengurangan")
print(x-y)
print(np.subtract(x,y))
```

```python
print("pengalian")
print(x*y)
print(np.multiply(x,y))
```

```python
print("pembagian")
print(x/y)
print(np.divide(x,y))
```
