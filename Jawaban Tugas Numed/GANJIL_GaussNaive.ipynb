```python
import numpy as np
```

```python
# input ukuran matriks dari user
m = int(input("masukkan jumlah persamaan (m): "))
n = int(input("masukkan jumlah variabel (n): "))

# generate matriks koefisien a dan vektor b secara acak rentang -50 sampai 50
a = np.random.uniform(-50.0, 50.0, (m, n))
b = np.random.uniform(-50.0, 50.0, m)

# tampilkan spl yang terbentuk
print("\nsistem persamaan linear (spl):")
for i in range(m):
    persamaan = ""
    for j in range(n):
        if j == 0:
            persamaan += "%.4f*x%d" % (a[i, j], j+1)
        else:
            if a[i, j] >= 0:
                persamaan += " + %.4f*x%d" % (a[i, j], j+1)
            else:
                persamaan += " - %.4f*x%d" % (abs(a[i, j]), j+1)
    persamaan += " = %.4f" % b[i]
    print(persamaan)

# tampilkan matriks augmentasi [a | b]
print("\nmatriks augmentasi [a | b]:")
print(np.column_stack((a, b)))
```

```python
# proses eliminasi gauss naif
# loop tiap kolom
iterasi = 0
for k in range(n):
    # cek pivot (diagonal utama matriks) apakah 0? 
    if a[k,k] == 0:
        # kalau iya, coba tukar baris
        for i in range(k, len(b)):      # cek baris di bawah
            if a[i, k] != 0:            # cari yang tidak nol         
                a[[k, i]] = a[[i, k]]   # tukar baris k (yang nol) dgn baris i (yang tdk nol) utk matriks a
                b[[k, i]] = b[[i, k]]   # tukar baris k (yang nol) dgn baris i (yang tdk nol) utk matriks b
                break

    # ambil pivot    
    pivot = a [k, k]

    # jika setelah swap pivot masih 0 berarti matriks tidak bisa diselesaikan
    if pivot == 0:  
        raise ValueError("Sistem tidak memiliki solusi unik atau tak terdefinisi.")
    
    # ubah pivot jadi 1
    a[k,:] /= pivot
    b[k] /= pivot
    
    # buat semua elemen di bawah pivot jadi 0
    for i in range(k+1, n):
        factor = a[i, k] 
        a[i, k:] -= factor * a[k, k:]
        b[i] -= factor * b[k]

    iterasi += 1

    print("\niterasi ke-%d (setelah eliminasi kolom %d):" % (iterasi, k+1))
    print(np.column_stack((a, b)))

# substitusi mundur untuk mendapatkan hasil (cari nilai x dari bawah ke atas)
print("\nmatriks a")
print(a)
print("\nmatriks b")
print(b)

# buat array solusi 
x = np.zeros(n)

# mulai dari variabel terakhir (z)
x[-1] = b[-1] / a[-1, -1]

# loop dari bawah ke atas
for i in range(n-2, -1, -1):
    # hitung jumlah a[i][j] * x[j] untuk j di kanan diagonal
    sum_val = b[i] - np.dot(a[i, i+1:], x[i+1:])
    # cari nilai x[i]
    x[i] = sum_val / a[i, i]

print("\nSolusi:")
for i in range(len(x)):
    print("x%d = %.4f" % (i+1, x[i]))
```
