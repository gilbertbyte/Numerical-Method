```python
import numpy as np
```

```python
# input ukuran matriks dari user
m = int(input("masukkan jumlah persamaan (m): "))
n = int(input("masukkan jumlah variabel (n): "))

# generate matriks koefisien a dan vektor b secara acak rentang -50 sampai 50
a = np.random.uniform(-50.0, 50.0, (m, n))
b = np.random.uniform(-50.0, 50.0, m)

# tampilkan spl yang terbentuk
print("\nsistem persamaan linear (spl):")
for i in range(m):
    persamaan = ""
    for j in range(n):
        if j == 0:
            persamaan += "%.4f*x%d" % (a[i, j], j+1)
        else:
            if a[i, j] >= 0:
                persamaan += " + %.4f*x%d" % (a[i, j], j+1)
            else:
                persamaan += " - %.4f*x%d" % (abs(a[i, j]), j+1)
    persamaan += " = %.4f" % b[i]
    print(persamaan)

# tampilkan matriks augmentasi [a | b]
print("\nmatriks augmentasi [a | b]:")
print(np.column_stack((a, b)))
```

```python
# proses eliminasi gauss jordan
# loop kolom (loop pivot)
iterasi = 0
for k in range(n):
    # swap kalau pivot 0
    if a[k, k] == 0:
        for i in range(k, m):
            if a[i, k] != 0:
                a[[k, i]] = a[[i, k]]
                b[[k, i]] = b[[i, k]]
                break

    # ambil pivot
    pivot = a[k, k]

    # jika setelah swap pivot masih 0 berarti matriks tidak bisa diselesaikan
    if pivot == 0:
        raise ValueError("sistem tidak memiliki solusi unik atau tak terdefinisi.")

    # ubah pivot jadi 1
    a[k, :] /= pivot
    b[k] /= pivot

    # loop semua baris (0 sampai m-1) untuk eliminasi kolom pivot
    for i in range(m):
        # skip baris pivot atau kalau sudah 0 tidak perlu dieliminasi
        if (i == k or a[i, k] == 0):
            continue
        # nilai pengali untuk menghilangkan elemen di kolom pivot
        factor = a[i, k]
        # loop setiap kolom dari kolom pivot sampai akhir
        for j in range(k, n):
            # operasi eliminasi gauss-jordan
            a[i, j] -= factor * a[k, j]     # matriks a
        b[i] -= factor * b[k]               # vektor b

    iterasi += 1
    print("\niterasi ke-%d (setelah eliminasi kolom %d):" % (iterasi, k+1))
    print(np.column_stack((a, b)))
```

```python
# tampilkan matriks akhir setelah eliminasi gauss jordan (bentuk identitas)
print("matriks augmentasi setelah eliminasi gauss jordan (bentuk identitas):")
print(np.column_stack((a, b)))

# solusi akhir langsung dari vektor b
print("\nsolusi akhir:")
for i in range(n):
    print("x%d = %.4f" % (i+1, b[i]))
```
