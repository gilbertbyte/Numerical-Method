```python
import numpy as np
import matplotlib.pyplot as plt
import math

def sin_taylor(x, n_terms=10):
    hasil = 0
    list_hasil = []
    
    for n in range(n_terms):
        term = ((-1)**n) * (x**(2*n+1)) / math.factorial(2*n+1)
        hasil += term
        list_hasil.append(hasil)
    
    return list_hasil
```

```python
sin1 = sin_taylor(np.pi / 2)
plt.plot(sin1, marker="o")
plt.ylabel("Nilai cos")
plt.xlabel("Loop ke berapa")
plt.grid()
plt.title("sin(pi/2)")
```

```python
sin2 = sin_taylor(np.pi / 4)
plt.plot(sin2, marker="o")
plt.ylabel("Nilai cos")
plt.xlabel("Loop ke berapa")
plt.grid()
plt.title("sin(pi/4)")
```
