```python
import numpy as np
import math

def rumus(x,n):
    atas = (-1) ** n
    bawah = math.factorial(2*n)
    samping = x**(2*n)
    hasil = atas/bawah * samping
    return hasil
```

```python
print(rumus(1,0))
print(rumus(1,1))
print(rumus(1,2))
```

```python
def hitung_cos(x):
    total = 0

    list_nilai = []
    for i in range(1,11):
        total += rumus(x, i)
        print(f"iterasi {i}")
        print(f"nilai cos {total}")
        list_nilai.append(total)

    return list_nilai
```

```python
nilai_y = hitung_cos(0)

import matplotlib.pyplot as plt

plt.plot(nilai_y, marker="o")
plt.ylabel("Nilai cos")
plt.xlabel("Loop ke berapa")
plt.grid()
plt.title("cos(0)")
```

```python
nilai_y = hitung_cos(np.pi/4)

import matplotlib.pyplot as plt

plt.plot(nilai_y, marker="o")
plt.ylabel("Nilai cos")
plt.xlabel("Loop ke berapa")
plt.grid()
plt.title("cos(pi/4)")
```

```python
def f(x):
    return np.sin(x) + np.log(x+2) - x ** 2 + np.cos(2*x)
```

```python
def biseksi(xi, xu, ea, es):
    list_error = []
    list_xr = []
    iter = 0
    while (ea > es):
        iter += 1
        print(f"Iterasi ke {iter}")
        print(f"xi : {xi:.4f}, xu: {xu:.4f}")
        print(f"f(xi) : {f(xi):.4f}, f(xu) : {f(xu):.4f}")
        xr = (xi+xu)/2
        print(f"xr : {xr:.4f}")

        if(iter > 1):
            ea = abs((xr - xr_old)/xr)*100
            print(f"ea : {ea:.4f}")
            list_error.append(ea)

        if(f(xi)*f(xr) < 0):
            xu = xr
        elif(f(xi)*f(xr)>0):
            xi = xr
        else:
            break
        list_xr.append(xr)
        xr_old = xr
    return list_error, list_xr
```

```python
list_error, list_xr = biseksi(0.5, 1.5, 100, 4)
```

```python
plt.figure(dpi=300)
plt.subplot(211)
plt.plot(range(1, 5), list_error, "r--", np.arange(1, 5), list_error, "r^")
plt.title("Nilai Error(es)")
plt.xticks(range(1,5))
plt.subplot(212)
plt.plot(range(1, 6), list_xr, "bo", np.arange(1, 6), list_xr, "b")
plt.title("Nilai Akar (Xr)")
plt.xticks(range(1,6))
plt.subplots_adjust(hspace=0.5)
plt.savefig('biseksi.png')
plt.show()
```
