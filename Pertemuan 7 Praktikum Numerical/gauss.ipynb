````python
import numpy as np
````

````python
# membuat matriks koefisien 
a = np.array([[2, 2, 1],
             [1, -1, 1],
             [3, 2, -2]], dtype=float)

# membuat vektor hasil
b = np.array([3,4,2], dtype=float)
# jumlah persamaan 
n = len(b)

# menampilkan matriks augmentasi [a | b]
print("Augmentasi Matriks: \n", np.column_stack((a, b)))

# eliminasi Gauss untuk membentuk matriks segitiga atas
# loop tiap kolom
for k in range(n):
    # cek pivot (diagonal utama matriks) apakah 0? 
    if a[k,k] == 0:
        # kalau iya, coba tukar baris
        for i in range(k, len(b)):      # cek baris di bawah
            if a[i, k] != 0:            # cari yang tidak nol         
                a[[k, i]] = a[[i, k]]   # tukar baris k (yang nol) dgn baris i (yang tdk nol) utk matriks a
                b[[k, i]] = b[[i, k]]   # tukar baris k (yang nol) dgn baris i (yang tdk nol) utk matriks b
                break

    # ambil pivot (diagonal utama)           
    pivot = a [k, k]

    # jika setelah swap pivot masih 0 berarti matriks tidak bisa diselesaikan
    if pivot == 0:  
        raise ValueError("Sistem tidak memiliki solusi unik atau tak terdefinisi.")
    
    # ubah semua pivot jadi 1 dengan membagi seluruh baris
    a[k,:] /= pivot
    b[k] /= pivot
    
    # buat semua elemen di bawah piot jadi 0
    for i in range(k+1, n):
        factor = a[i, k] 
        a[i, k:] -= factor * a[k, k:]
        b[i] -= factor * b[k]

# hasil setelah eliminasi
print("\nMatriks setelah eliminasi Gauss:")
print(np.column_stack((a, b)))

# substitusi mundur untuk mendapatkan hasil (cari nilai x dari bawah ke atas)
print("\nmatriks a")
print(a)
print("\nmatriks b")
print(b)

# buat array solusi 
x = np.zeros(n)

# mulai dari variabel terakhir (z)
x[-1] = b[-1] / a[-1, -1]

# loop dari bawah ke atas
for i in range(n-2, -1, -1):
    # hitung jumlah a[i][j] * x[j] untuk j di kanan diagonal
    sum_val = b[i] - np.dot(a[i, i+1:], x[i+1:])
    # cari nilai x[i]
    x[i] = sum_val / a[i, i]

print("\nSolusi:")
for i in range(len(x)):
    print("x%d = %.4f" % (i+1, x[i]))
````
