````python
import numpy as np
````

````python
# matriks koefisien
a = np.array([[0, 2, 3],
             [1, 4, 6],
             [2, 5, 7]], dtype=float)

# vektor hasil
b = np.array([5,9,12], dtype=float)
# jumlah variabel
n = len(b)

# gabung matriks a dan b
print("Augmentasi Matriks: \n", np.column_stack((a, b)))

# loop kolom
for k in range(n):
    # swap kalau pivot 0
    if a[k,k] == 0:
        for i in range(k, len(b)):
            if a[i, k] != 0:
                a[[k, i]] = a[[i, k]]
                b[[k, i]] = b[[i, k]]
                break
    # ambil pivot
    pivot = a [k, k]
    # kalau ttp 0 gbs dihitung
    if pivot == 0:  
        raise ValueError("Sistem tidak memiliki solusi unik atau tak terdefinisi.")
    # ubah pivot jd 1
    a[k, :] /= pivot
    b[k] /= pivot

    # loop semua baris (0 sampai n-1)
    for i in range(n):
        # skip baris pivot or kalau sudah 0 tdk perlu dieliminasi
        if (i == k or a[i, k] == 0): 
            continue
        # nilai pengali untuk menghilangkan elemen di kolom pivot
        factor = a[i, k]
        # loop setiap kolom, dari kolom pivot sampai akhir
        for j in range(k, n):
            # operasi eliminasi Gauss-Jordan
            a[i, j] -=  factor * a[k, j]        # matriks a
        b[i] -= factor * b[k]                   # vektor b

# tampilkan hasil matriks setelah eliminasi Gauss-Jordan
print("\n Gauss Jordan: \n", np.column_stack((a,b)))
print()

# tampilkan nilai vektor hasil b satu per satu
for i in range(len(b)):
    print("b%d = %.4f" % (i+1, b[i]))
````
